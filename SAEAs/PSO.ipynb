```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Ackley function with separate x and y arguments
def ackley_function(x, y, a=20, b=0.2, c=2 * np.pi):
    term1 = -a * np.exp(-b * np.sqrt(0.5 * (x**2 + y**2)))
    term2 = -np.exp(0.5 * (np.cos(c * x) + np.cos(c * y)))
    return term1 + term2 + a + np.exp(1)

# Particle class definition
class Particle:
    def __init__(self, bounds):
        # Initialize position as two coordinates (x and y)
        self.position = np.array([np.random.uniform(bounds[0][0], bounds[1][0]), 
                                  np.random.uniform(bounds[0][1], bounds[1][1])])
        # Initialize velocity as two components (vx and vy)
        self.velocity = np.random.uniform(-1, 1, 2)
        self.best_position = self.position.copy()
        self.best_score = ackley_function(self.position[0], self.position[1])
        
    def update_velocity(self, global_best_position, w=0.5, c1=1.5, c2=1.5):
        inertia = w * self.velocity
        cognitive = c1 * np.random.rand() * (self.best_position - self.position)
        social = c2 * np.random.rand() * (global_best_position - self.position)
        self.velocity = inertia + cognitive + social
    
    def update_position(self, bounds):
        self.position += self.velocity
        for i in range(2):  # Only two dimensions, x and y
            self.position[i] = np.clip(self.position[i], bounds[0][i], bounds[1][i])

# Particle Swarm Optimization with Visualization
def particle_swarm_optimization(func, bounds, num_particles=30, max_iter=100):
    particles = [Particle(bounds) for _ in range(num_particles)]
    
    # Initialize global best
    global_best_position = particles[0].position
    global_best_score = func(global_best_position[0], global_best_position[1])
    
    # Set up the Ackley function landscape plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    plot_ackley_function(ax, bounds)
    
    # Scatter plot for particles
    particle_scatter = ax.scatter([], [], [], color="red", s=50)
    
    # Optimization loop with visualization
    for iteration in range(max_iter):
        positions = np.array([particle.position for particle in particles])
        
        # Update global and personal bests
        for particle in particles:
            score = func(particle.position[0], particle.position[1])
            if score < particle.best_score:
                particle.best_score = score
                particle.best_position = particle.position.copy()
            if score < global_best_score:
                global_best_score = score
                global_best_position = particle.position.copy()
                
        for particle in particles:
            particle.update_velocity(global_best_position)
            particle.update_position(bounds)
        
        # Update particle positions in plot
        particle_scatter._offsets3d = (positions[:, 0], positions[:, 1], [func(pos[0], pos[1]) for pos in positions])
        plt.draw()
        plt.pause(0.1)  # Pause for visualization update

    plt.show()
    return global_best_position, global_best_score

# Function to plot the Ackley function surface
def plot_ackley_function(ax, bounds):
    x = np.linspace(bounds[0][0], bounds[1][0], 100)
    y = np.linspace(bounds[0][1], bounds[1][1], 100)
    X, Y = np.meshgrid(x, y)
    Z = ackley_function(X, Y)  # Apply the Ackley function to each point on the grid

    # Plot the Ackley function surface
    ax.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.6, edgecolor="k")
    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Ackley Function Value")
    ax.set_title("Ackley Function with Particle Movements")

# Define bounds and run the PSO
bounds = [(-5, -5), (5, 5)]
best_position, best_score = particle_swarm_optimization(ackley_function, bounds, num_particles=30, max_iter=100)

print("Best Position:", best_position)
print("Best Score:", best_score)
```
